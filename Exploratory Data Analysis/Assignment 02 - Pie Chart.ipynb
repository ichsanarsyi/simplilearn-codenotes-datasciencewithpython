```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cause = "Chronic Diseases", "Unintentional Injuries", "Alzheimers", "Influenze and Pnemumonia", "Sepsis", "Others"
percentile = [62, 5, 4, 2, 1, 26]
```

```python
plt.figure(figsize=(10,10))
explode = (0.05, 0, 0, 0, 0, 0)
plt.pie(percentile, labels=cause, explode=explode, autopct="%1.1f%%", startangle=70)
plt.axis("equal")
plt.title("Ohio State 2012: Leading Causes of Death")
plt.show()
```
