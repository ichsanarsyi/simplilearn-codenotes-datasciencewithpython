```python
# Scipy.stats poisson class is used along with pmf method to calculate the value of probabilities
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
# Random variable representing number of restaurants
# Mean number of occurencs of restaurants in 10 miles is 2

x = [0, 1, 2, 3, 4, 5]
lmbda = 2
```

```python
# probabilitiy values
poisson_pd = poisson.pmf(x, lmbda)
```

```python
# Plot the probability distribution

fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(x, poisson_pd, "bo", ms=3, label="poisson pmf")
plt.ylabel("Probability", fontsize="18")
plt.xlabel("x = No. of Rstaurants", fontsize="18")
plt.title("Poisson Distribution - o. of Restaurants Vs Probability", fontsize="18")
ax.vlines(x, 0, poisson_pd, colors="b", lw=5, alpha=0.5)
```
